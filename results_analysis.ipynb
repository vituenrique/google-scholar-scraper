```python
import pandas as pd
```

```python
df_all_papers = pd.read_json("./papers.json")
df_all_papers
```

```python
df_papers_by_citation = pd.read_csv("./papers_by_citation.csv")
df_papers_by_citation
```

### Encontrar distribuição de artigos com melhor média de citações por ano
#### Top 10% - 2 artigos por ano

```python
df_papers_by_citation["cit/year"].describe(percentiles=[0.25, 0.5, .75, .9])
```

#### Filtrando top 10% artigos

```python
df_top_10 = df_papers_by_citation[df_papers_by_citation["cit/year"] >= 2]
df_top_10
```

#### Top 10% - Distribuição por ano

```python
df_top_10["Year"].value_counts()
```

#### Top 10% - Distribuição por Publisher

```python
df_top_10["Publisher"].value_counts()
```

#### Top 10% - Distribuição por Evento

```python
df_top_10["Venue"].value_counts()
```

```python
import re
df_all_papers["title"] = df_all_papers["title"].apply(lambda title: re.sub("[\(\[].*?[\)\]]", "", title).strip())
df_all_papers
```

```python
final_papers = pd.merge(df_top_10, df_all_papers, how="inner", left_on="Title", right_on="title")
final_papers = final_papers.drop(columns=["title", "Rank", "title_link", "publication_info", "pdf_link"])
final_papers
```

```python
final_papers.to_csv("./best_papers.csv")
```
