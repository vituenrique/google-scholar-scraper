```python
import pandas as pd
```

```python
df_papers = pd.read_json("./papers.json")
df_papers["year"] = df_papers["year"].astype("Int64")
df_papers
```

### Encontrar distribuição de artigos com melhor média de citações por ano
#### Top 10% - 2 artigos por ano

```python
df_papers["cit/year"].describe(percentiles=[0.25, 0.5, .75, .9, .95])
```

#### Filtrando top 10% artigos

```python
df_top_10 = df_papers[df_papers["cit/year"] >= 10].reset_index(drop=True)
df_top_10 = df_top_10.sort_values(by="cit/year", ascending=False, ignore_index=True)
df_top_10
```

#### Top 10% - Distribuição por ano

```python
df_papers["year"].value_counts()
```

```python
df_top_10["year"].value_counts()
```

#### Top 10% - Distribuição por Publisher

```python
df_papers["publisher"].value_counts()
```

```python
df_top_10["publisher"].value_counts()
```

#### Top 10% - Distribuição por Evento

```python
df_top_10["vanue"].value_counts()
```

```python
import re
df_top_10["title"] = df_top_10["title"].apply(lambda title: re.sub("[\(\[].*?[\)\]]", "", title).strip())
df_top_10
```
